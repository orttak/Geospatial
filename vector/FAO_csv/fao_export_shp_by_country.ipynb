```python
import pandas as pd
import geopandas as gpd
```

```python
raw_countries=gpd.read_file('../../World_Countries/countries_border.shp')
```

```python
raw_countries.head()
```

```python
brazil=raw_countries[raw_countries['GID_0']=='BRA']
```

```python
polinator_brazil=pd.read_csv('fao/polinator_sample.csv')
```

```python
polinator_brazil.head()
```

```python
den=pd.merge(brazil,polinator_brazil,how='inner', left_on='GID_0', right_on='Country')
den.to_file('fao/polinator_brasil.shp',driver='ESRI Shapefile')
```
